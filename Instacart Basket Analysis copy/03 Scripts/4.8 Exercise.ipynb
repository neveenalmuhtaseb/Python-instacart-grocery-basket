## Aggregating data with agg()

```python
#  calculate the mean of the “order_number” column grouped by the “department_id” column
df_sub.groupby('department_id').agg({'user_order_number': ['mean']})
# or simply df.groupby('department_id')['order_number'].mean()
```

## Aggregating data with transform()

```python
# create loyalty flag column,first create aggregated column
df['max_order'] = df.groupby(['user_id'])['user_order_number'].transform(np.max)
```

```python
df.head(15)
```

```python
# create flag column using loc()
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df['loyalty_flag'].value_counts(dropna=False)
```

```python
# create spending flag column,first create aggregated column
df['prices_average'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# create spending flag column using loc()
df.loc[df['prices_average'] >= 10, 'spending_flag'] = 'High Spender'
df.loc[df['prices_average'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df['spending_flag'].value_counts(dropna=False)
```

```python
# determine frequent versus non-frequent customers.
df['prior_order_median'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# create flag column using loc()
df.loc[df['max_order'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
df.loc[(df['max_order'] <= 20) & (df['max_order'] > 10), 'frequency_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
